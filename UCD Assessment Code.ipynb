```python
#Importing Python libraries:
#Let's import Python libraries for analysis.
import pandas as pd
import numpy as np
import ast

#Let's import Python libraries for plotting.
import seaborn as sns
import matplotlib.pyplot as plt

#Let's import Python libraries for calling an API.
import requests



#Let's import the datasets for analysis and convert them into dataframes.
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv') 



#Before we move forward with the analysis, let's connect to an api to get information of a movie at random.
url = "http://www.omdbapi.com/?i=tt3896198&apikey=987bc186"
api = requests.get(url)
print(api)
json_data = api.json()
for key, value in json_data.items():
    print(key + ':', value)
print(json_data['Title'])



#Let's go back to our dataframes with which we wil be looking into for our project.

#Let's have a look at our data prior merging.
print('MOVIES & CREDITS DATAFRAMES PRIOR MERGING')

#Information from the dataframes:

#Information from the movies dataframe:
print('MOVIES DATAFRAME INFORMATION')
print(np.shape(movies))
print(movies.info())
print(movies.head())


# Let's confirm that the id column is the primary key: there is one unique id for each row.
print(len(movies.id.unique()))


#Information from the credits dataframe:
print('CREDITS DATAFRAME INFORMATION')
print(np.shape(credits))
print(credits.info())
print(credits.head())

# Let's confirm that the id column is the primary key: there should be one unique id for each row.
print(len(credits.movie_id.unique()))



# Let's count the unique values in each column.
movies_id_count = pd.unique(movies.id)
movies_title_count = pd.unique(movies.title)
credits_count = pd.unique(credits.movie_id)
credits_title_count = pd.unique(credits.title)

# Let's initialize the variables to 0 for counting:
movies_unique = 0
movies_title_unique = 0
credits_unique = 0
credits_title_unique = 0

# Let's utilize a for loop of each group to count for unique ids and titles.
for x in movies_id_count:
    movies_unique += 1
    
for x in movies_title_count:
    movies_title_unique += 1

for x in credits_count:
    credits_unique += 1
    
for x in credits_title_count:
    credits_title_unique += 1

print(f'There are {movies_unique} ids in movies dataframe.')
print(f'There are {movies_title_unique} titles in movies dataframe.')
print(f'There are {credits_unique} ids in credits dataframe.')
print(f'There are {credits_title_unique} titles in credits dataframe.')



#Before merging our data, let's clean it up first.

#Let's check for missing values:

missing_values_movies = movies.isnull().sum().values
print('The movies dataset does not have any null values.') if np.sum(missing_values_movies) == 0 else print('The movies dataset has null values.')

missing_values_credits = credits.isnull().sum().values
print('The credits dataset does not have any null values.') if np.sum(missing_values_credits) == 0 else print('The credits dataset has null values.')

#There are missing values in the movies dataset, let's find out where the missing values are:
movies.isnull().sum().plot(kind ="bar", title = "Missing values in movies dataset",color="red")
plt.show()


#Let's check for N/A values:

na_values_movies = movies.isna().sum()
na_movies = na_values_movies.to_numpy()
print('The movies dataset does not have any N/A values.') if np.sum(na_movies) == 0 else print('The movies dataset has N/A values.')

na_values_credits = credits.isna().sum()
na_credits = na_values_credits.to_numpy()
print('The credits dataset does not have any N/A values.') if np.sum(na_credits) == 0 else print('The credits dataset has N/A values.')

#There are N/A values in the movies dataset, let's find out where the missing values are:
movies.isna().sum().plot(kind = "bar", title = "N/A values in movies dataset",color="blue")
plt.show()


#Let's check for duplicate values:

drop_duplicates = movies.drop_duplicates(subset = ["id","title"])
print('The movies dataset does not have any duplicated values.') if movies.shape == drop_duplicates.shape else print('The movies dataset has duplicated values.')

drop_duplicates = credits.drop_duplicates(subset = ["movie_id","title"])
print('The credits dataset does not have any duplicated values.') if credits.shape == drop_duplicates.shape else print('The credits dataset has duplicated values.')


#Unfortunately, there are missing and N/A values in the movies dataset. 
#We could drop the missing and N/A values but that would mean losing a lot of observations.
#Let's replace these values with 0 so that we do not lose any data:

cleaned_movies = movies.fillna(0)

#Let's check again if the movies dataset is now cleaned:

missing_values_cleaned_movies = cleaned_movies.isnull().sum().values
print('The cleaned movies dataset does not have any null values.') if np.sum(missing_values_cleaned_movies) == 0 else print('The cleaned movies dataset has null values.')

na_values_cleaned_movies = cleaned_movies.isna().sum()
na_cleaned_movies = na_values_cleaned_movies.to_numpy()
print('The cleaned movies dataset does not have any N/A values.') if np.sum(na_cleaned_movies) == 0 else print('The cleaned movies dataset has N/A values.')

#Data is clean now!



#Let's merge the dataframes into our the dataset we will be working for our analysis.

ucd_df = cleaned_movies.merge(credits, how = 'inner', left_on = 'id', right_on = 'movie_id')
print(ucd_df.head())
print(np.shape(ucd_df))

#Let's drop the duplicate columns from our final dataset.

ucd_df = ucd_df.drop(['homepage', 'original_title', 'movie_id', 'title_y'], axis = 1)
print(ucd_df.head())
print(np.shape(ucd_df))

#Let's rename our columns:

ucd_df = ucd_df.rename({'id':'movie_id', 'title_x':'title'}, axis = 1)
list(ucd_df)



#Before moving forward with the analysis, we need to convert data from our columns into a more usable format.

#Let's copy the crew column so we can have two columns with crew and director information:
ucd_df['director'] = ucd_df['crew'].copy()


#Let's extract genre names, reduce the number of genres to one, and clean up the output:
def convert_genre(text):
    genres = []  
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 1:
            genres.append(i['name'])
        counter+=1
    return genres
ucd_df['genres'] = ucd_df['genres'].apply(convert_genre)
ucd_df['genres'] = ucd_df['genres'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['genres'] = ucd_df['genres'].str.replace("'", "") #To remove the inverted brackets from the results.


#Let's extract keywords names, reduce the number of keywords to one, and clean up the output:
def convert_keywords(text):
    keywords = []  
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 1: 
            keywords.append(i['name'])
        counter += 1
    return keywords
ucd_df['keywords'] = ucd_df['keywords'].apply(convert_keywords)
ucd_df['keywords'] = ucd_df['keywords'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['keywords'] = ucd_df['keywords'].str.replace("'", "") #To remove the inverted brackets from the results.


#Let's extract Production Company names, reduce the number of companies to one, and clean up the output:
def convert_production_companies(text):
    production_companies = []  
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 1:
            production_companies.append(i['name'])
        counter += 1
    return production_companies
ucd_df['production_companies'] = ucd_df['production_companies'].apply(convert_production_companies)
ucd_df['production_companies'] = ucd_df['production_companies'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['production_companies'] = ucd_df['production_companies'].str.replace("'", "") #To remove the inverted brackets from the results.


#Let's extract production countries data and clean up the output:
def convert_production_countries(text):
    production_countries = []  
    for i in ast.literal_eval(text):
        production_countries.append(i['name'])
    return production_countries
ucd_df['production_countries'] = ucd_df['production_countries'].apply(convert_production_countries)
ucd_df['production_countries'] = ucd_df['production_countries'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['production_countries'] = ucd_df['production_countries'].str.replace("'", "") #To remove the inverted brackets from the results.


#Let's convert the release date column to release year to simplify our analysis:
ucd_df['release_date'] = pd.to_datetime(ucd_df['release_date']).dt.year


#Let's extract spoken languages and clean up the output:
def convert_spoken_languages(text):
    spoken_languages = []  
    for i in ast.literal_eval(text):
        spoken_languages.append(i['name'])
    return spoken_languages
ucd_df['spoken_languages'] = ucd_df['spoken_languages'].apply(convert_spoken_languages)
ucd_df['spoken_languages'] = ucd_df['spoken_languages'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['spoken_languages'] = ucd_df['spoken_languages'].str.replace("'", "") #To remove the inverted brackets from the results.


#Let's extract cast name and clean up the output:
def convert_cast(text):
    cast = []  
    for i in ast.literal_eval(text):
        cast.append(i['name'])
    return cast
ucd_df['cast'] = ucd_df['cast'].apply(convert_cast)
ucd_df['cast'] = ucd_df['cast'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['cast'] = ucd_df['cast'].str.replace("'", "") #To remove the inverted brackets from the results.
ucd_df['cast'] = ucd_df['cast'].str.extract('(.*?),', expand=False) #To extract the first name listed in the row.


#Let's extract crew gender:
def convert_crew(text):
    crew = []  
    for i in ast.literal_eval(text):
        crew.append(i['gender'])
    return crew
ucd_df['crew'] = ucd_df['crew'].apply(convert_crew)

#Crew gender is identified by the following integers: 0 = Not Known, 1 =Female, 2 = Male.
gender_dictionary = {0:'Not known', 1:'Female', 2:'Male'}

#Let's convert these integers in the crew column into values from our dictionary and clean up the results:
ucd_df['crew'] = [[gender_dictionary[k] for k in lst] for lst in ucd_df['crew']]
ucd_df['crew'] = ucd_df['crew'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['crew'] = ucd_df['crew'].str.replace("'", "") #To remove the inverted brackets from the results.


#Let's extract the director's name and clean up the output:
def fetch_director(text):
    director = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            director.append(i['name'])
    return director 
ucd_df['director'] = ucd_df['director'].apply(fetch_director)
ucd_df['director'] = ucd_df['director'].astype(str).str.strip('[]') #To remove the brackets from the results.
ucd_df['director'] = ucd_df['director'].str.replace("'", "") #To remove the inverted brackets from the results.



#Given that we have performed some extraction activities, let's clean up our data one last time.

#Let's check for missing values:
missing_values_ucd = ucd_df.isnull().sum().values
print('The UCD dataframe does not have any null values.') if np.sum(missing_values_ucd) == 0 else print('The UCD dataframe has null values.')

#Let's check for N/A values:
na_values_ucd = ucd_df.isna().sum()
na_ucd = na_values_ucd.to_numpy()
print('The UCD dataframe does not have any N/A values.') if np.sum(na_ucd) == 0 else print('The UCD dataframe has N/A values.')

#Unfortunately, there are missing and N/A values in the UCD dataframe. 
#We could drop the missing and N/A values but that would mean losing a lot of observations.
#Let's replace these values with 0 so that we do not lose any data:
ucd_df = ucd_df.fillna(0)



#Let's rename our columns for one last time so we can start our analysis:
ucd_df = ucd_df.rename({'genres':'genre', 'release_date':'release_year', 'crew':'crew_gender'}, axis=1)
list(ucd_df)
print(ucd_df.head())



#Let's start with our data analysis:


#Let's have a look at the number of movies released by year. First, we will want to group our data by release year:
movies_released_by_year = ucd_df.groupby("release_year")["title"].count()
movies_released_by_year.plot(kind = "line", title = 'Movies released by year', rot = 45, color = "magenta")
plt.xlabel('Release years')
plt.ylabel('Movie count')
plt.show()

#To understand the data better, let's have a look at the movie identified as the first movie released in our source:
print(movies_released_by_year.head())

#Let's have a look the movie identified as the last release recorded in our source:
movies_released_by_year_descending = ucd_df.sort_values(by = "release_year", ascending = False)
print(movies_released_by_year_descending['release_year'].head())

#Let's have at the release year with more results:
most_movies_released_by_year = movies_released_by_year.sort_values(ascending = False)
print(most_movies_released_by_year.head())


#Let's have a look at movie runtimes distribution:
plt.hist(ucd_df['runtime'], bins = 5, orientation = "horizontal", color = "pink", edgecolor = "white")
plt.title('Movie runtimes')
plt.xlabel('Movie count')
plt.ylabel('Runtime')
plt.show()


#Let's have a look at average vote distribution:
plt.hist(ucd_df['vote_average'], bins = 5, orientation = "horizontal", color = "purple", edgecolor = "white")
plt.title('Movies by average vote')
plt.xlabel('Movie count')
plt.ylabel('Average vote')
plt.show()


#Let's have a look at the number of movies by genre:
sns.catplot(x = "genre", data = ucd_df, kind = "count")
plt.title('Count of movies per genre')
plt.xlabel('Movie genres')
plt.ylabel('Movie count')
plt.xticks(rotation = 90)
plt.show()


#Let's have a look at budget and revenue data by genre:
value_counts = ucd_df['genre'].value_counts()
top_10_genres = value_counts.head(10).index.tolist()
top_10_genres_df = ucd_df[ucd_df['genre'].isin(top_10_genres)]

sns.scatterplot(x = 'budget', y = 'revenue', data = top_10_genres_df, hue = 'genre', alpha = 0.75)
plt.title('Movie budget/revenue relationship by genre')
plt.xlabel('Movie budget')
plt.ylabel('Movie revenue')
plt.xticks(rotation = 45)
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
plt.show()


#Let's get a top ranking of this dataframe directors:
value_counts = ucd_df['director'].value_counts()
top_10_director = value_counts.head(10).index.tolist()
top_10_director_df = ucd_df[ucd_df['director'].isin(top_10_director)]

#Let's have a look at the count of movies per director in descending order:
movie_count_per_director = top_10_director_df.groupby("director")["title"].count()
movie_count_per_director_sorted = movie_count_per_director.sort_values(ascending = False)
sort = ["Steven Spielberg",
        "Woody Allen",
        "Clint Eastwood", 
        "Martin Scorsese", 
        "Ridley Scott",
        "Spike Lee", 
        "Renny Harlin",
        "Steven Soderbergh",
        "Oliver Stone"]

print(movie_count_per_director_sorted)

sns.catplot(x = "director", data = top_10_director_df, kind = "count", order = sort)
plt.title('Count of movies per director')
plt.xlabel('Directors')
plt.ylabel('Movie count')
plt.xticks(rotation = 90)
plt.plot()


#Let's plot a pairplot for additional insights based by movie popularity:
sns.pairplot(ucd_df, hue = 'popularity')
plt.suptitle('Movie popularity correlation', fontsize = 20, y = 1.03)
plt.show()
```
